# Cropping Mercury frames

```python
# List to hold the cut data of each mercury frame
raw_mercury_cut = []

# Define the y-axis ranges for each frame
y_ranges_raw_hg = [(24, 94), (24, 94)]

# Loop through each corrected frame, cut along y-axis, and extract the data
for i in range(2):
    # Get the full frame data
    mercury_data = raw_mercury[i]

    # Cut the frame data along the y-axis according to the specified range
    hgy_start, hgy_end = y_ranges_raw_hg[i]
    cut_mercury_data = mercury_data[hgy_start:hgy_end, :]

    # Append the cut frame data to the list
    raw_mercury_cut.append(cut_mercury_data)
```

```python
# Step 1: Sum along the y-axis for each mercury frame
summed_mercury = []

for frame in raw_mercury_cut:
    summed_1d = frame.sum(axis=0)  # Sum along the y-axis
    summed_mercury.append(summed_1d)

# Step 2: Average the summed mercury spectra across the frames
average_mercury_spectrum = np.mean(summed_mercury, axis=0)

# Step 3: Plot the averaged mercury spectrum
plt.figure(figsize=(10, 5))
plt.plot(average_mercury_spectrum, label="Averaged Mercury Spectrum")
plt.title("Averaged 1D Mercury Spectrum")
plt.xlabel("Pixel Position")
plt.ylabel("Summed Intensity")
plt.legend()
plt.show()
```

```python
#Wavelengths of identified Mercury lines
#from https://www.newport.com/medias/sys_master/images/images/h55/hfd/8797293281310/Typical-Spectra-of-Spectral-Calib-Lamps.pdf

identified_hg_lines = {
    "Hg 435.86 nm": 435.86,
    "Hg 407.78 nm": 407.78,
    "Hg 404.66 nm": 404.66
}

hg_wavelength=sorted(identified_hg_lines.values())
hg_dict=sorted(identified_hg_lines.keys())
```

# Calibrating Pixel Position into Wavelength in nm

```python
#Finding the peaks of each observed Mercury line

pixel_pos=np.arange(len(average_mercury_spectrum))

peaks, _ = find_peaks(average_mercury_spectrum, prominence=26000)
peak_points= pixel_pos[peaks]

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(pixel_pos, average_mercury_spectrum)
plt.plot(peak_points, average_mercury_spectrum[peaks], "x", label='Detected Peaks', color='red')
plt.title('Mercury Light Spectrum')
plt.xlabel('Pixel Position')
plt.ylabel('Intensity')
plt.legend()
plt.grid()

plt.show()
```

```python
#Finding the pixel to wavelength ratio

w_coef = np.polyfit(peak_points,hg_wavelength,1)
poly1d_w = np.poly1d(w_coef)

plt.plot(peak_points, hg_wavelength, 'bo', peak_points, poly1d_w(peak_points))
plt.xlabel('Peaks Pixel Position')
plt.ylabel('Mercury Wavelenght (nm)')
plt.show()
```

```python
calibrated_wave=(pixel_pos*w_coef[0])+w_coef[1]

plt.figure(figsize=(12, 6))
plt.plot(calibrated_wave, average_mercury_spectrum)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity')
plt.grid()

# Annotate Neon lines
for i in range(len(hg_wavelength)):
    #plt.axvline(x=n_wavelength[i], ymin=0.1, color='red', linestyle='--')  # Vertical line for each emission line
    plt.text(hg_wavelength[i],average_mercury_spectrum[peaks][i]/1.25, hg_dict[i],
             fontsize=10, rotation=270, color='black', ha='left')
```

```python
# Constants
h = 6.62607015e-34  # Planck's constant (Js)
c = 3.0e8           # Speed of light (m/s)
k_B = 1.380649e-23  # Boltzmann constant (J/K)

# Define the Planck function (Blackbody spectrum)
def blackbody_spectrum(wavelength, T):
    """Planck's law for blackbody radiation."""
    wavelength_m = wavelength * 1e-9  # Convert from nm to m
    return (2 * h * c**2) / (wavelength_m**5) * (1 / (np.exp((h * c) / (wavelength_m * k_B * T)) - 1))

# Function to detect and mask large dips (based on depth and width)
def detect_and_mask_large_dips(wavelength, intensity, min_dip_depth=200, min_dip_width=5):
    """
    Detects large dips in the spectrum and masks them.

    Parameters:
        wavelength: Array of wavelengths (nm)
        intensity: Array of intensity values corresponding to the wavelengths
        min_dip_depth: Minimum depth of dips (intensity drop) to be considered as a dip
        min_dip_width: Minimum number of consecutive points to be considered a dip

    Returns:
        masked_intensity: The intensity array with large dips masked out (set to np.nan)
        mask: A boolean array where True represents a masked (dipped) region
    """
    mask = np.zeros_like(intensity, dtype=bool)
    masked_intensity = np.copy(intensity)

    # Loop through the intensity array and find large dips
    for i in range(1, len(intensity) - 1):
        # Check for local minima (dips)
        if intensity[i] < intensity[i-1] and intensity[i] < intensity[i+1]:
            # Check if the dip depth is significant (deeper than min_dip_depth)
            dip_depth = intensity[i-1] - intensity[i]
            if dip_depth >= min_dip_depth:
                # Expand the dip region to include nearby points, if necessary
                left_idx = i - min_dip_width if i - min_dip_width >= 0 else 0
                right_idx = i + min_dip_width if i + min_dip_width < len(intensity) else len(intensity)

                # Mask out this region
                mask[left_idx:right_idx] = True
                masked_intensity[left_idx:right_idx] = np.nan  # Or 0, depending on your preference

    return masked_intensity, mask

# Function to mask known emission dips by wavelength range
def mask_known_emission_dips(wavelength, intensity, dip_ranges):
    """
    Masks known emission dips by wavelength range.

    Parameters:
        wavelength: Array of wavelengths (nm)
        intensity: Array of intensity values
        dip_ranges: List of tuples defining wavelength ranges to mask (e.g., [(430, 440), (410, 420)])

    Returns:
        masked_intensity: The intensity array with known emission dips masked
        mask: A boolean array where True represents a masked region
    """
    mask = np.zeros_like(intensity, dtype=bool)
    masked_intensity = np.copy(intensity)

    for (start, end) in dip_ranges:
        # Identify indices that fall within each dip range
        dip_mask = (wavelength >= start) & (wavelength <= end)
        mask[dip_mask] = True
        masked_intensity[dip_mask] = np.nan  # Or 0, depending on your preference

    return masked_intensity, mask

# Assuming master_etatau_1d and calibrated_wave are already defined
# Define known emission dip ranges in wavelength (nm)
dip_ranges = [(375.3, 377), (378.3, 380), (382, 384), (387, 390), (395, 399), (408, 412), (432, 436)]  # Modify as needed based on your data
#dip_ranges = [(374.9, 375.5), (377.2, 378.3), (380, 382), (384, 387), (390, 395), (399, 408), (412, 432), (436, 445)]

# Step 1: Mask known emission dips
masked_intensity, mask_known = mask_known_emission_dips(calibrated_wave, master_etatau_1d, dip_ranges)

# Step 2: Mask additional large dips based on depth and width
masked_intensity, mask_large_dips = detect_and_mask_large_dips(calibrated_wave, masked_intensity, min_dip_depth=200, min_dip_width=5)

# Step 3: Fit the blackbody model to the masked data
def fit_blackbody_spectrum(wavelength, intensity):
    initial_guess = [5000]  # Initial guess for temperature (in Kelvin)

    # Perform the fit using curve_fit, ignoring the masked regions
    popt, pcov = curve_fit(blackbody_spectrum, wavelength[~mask_known & ~mask_large_dips], intensity[~mask_known & ~mask_large_dips], p0=initial_guess)

    # Extract the fitted temperature
    T_fitted = popt[0]
    T_fitted_err = np.sqrt(np.diag(pcov))[0]  # Uncertainty in the temperature

    return T_fitted, T_fitted_err, popt, pcov

# Fit the blackbody spectrum to the masked intensity
T_fitted, T_fitted_err, popt, pcov = fit_blackbody_spectrum(calibrated_wave, masked_intensity)

# Generate the fitted blackbody spectrum with the fitted temperature
fitted_spectrum = blackbody_spectrum(calibrated_wave, T_fitted)

# Plot the original data, the masked data, and the fitted blackbody spectrum
plt.figure(figsize=(10, 6))
plt.plot(calibrated_wave, master_etatau_1d, label='Original Spectrum', color='blue')
plt.plot(calibrated_wave, masked_intensity, label='Masked Spectrum (Dips Removed)', color='orange', linestyle='--')
plt.plot(calibrated_wave, fitted_spectrum, label=f'Fitted Blackbody Spectrum\nT={T_fitted:.0f} K', color='red', linestyle='--')
plt.title("Blackbody Fit to Emission Spectrum of Eta Tau (Large Dips Masked)")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Intensity")
plt.legend()
plt.show()

# Print the fitted temperature and its uncertainty
print(f"Fitted Temperature: {T_fitted:.0f} K  {T_fitted_err:.0f} K")
```
